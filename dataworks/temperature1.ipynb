### 기온 데이터 분석

```python
# 파일입출력, csv 모듈
import csv
import matplotlib.pyplot as plt
# with open("datas/temp_seoul.csv", "r", encoding-"utf-8") as f:
with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f) # 데이터 읽어온 자료 저장
    # next(data)
    header = next(data)
    # print(header)
    # for i in data:
    #     print(i)

    # 최고 기온 데이터를 숫자 타입으로 변환
    temp_high=[]
    for row in data:
        if row[-1] !='':
            if int(row[0].split('-')[0])>= 1980:
                temp_high.append(float(row[-1]))
            
plt.rc('font', family='Malgun Gothic')
plt.title("1980년 이후 서울의 최고기온")
plt.plot(temp_high, color='r')
plt.show()
    
```
