```python
import matplotlib.pyplot as plt
import random as a

# plt.hist([1, 1, 2, 3, 4, 5, 5, 5, 5, 6])
dice = []
for i in range(10):
    dice.append(a.randint(1, 6))
print(dice)
plt.hist(dice, bins=6)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f) 
    header = next(data)

    temp_high=[]
    for row in data:
        if row[-1] !='':
            if int(row[0].split('-')[0])>= 1980:
                temp_high.append(float(row[-1]))
            
plt.rc('font', family='Malgun Gothic')
plt.title("1980년 이후 서울의 최고기온")
plt.hist(temp_high, bins=100, color='r')
plt.xlabel('온도')
plt.ylabel('빈도수')
plt.show()
    
```
