### 시각화 라이브러리 사용 - matplotlib

```python
import matplotlib.pyplot as plt
# plt.plot([1, 2, 3 ,4])
plt.title("Basic Gragh")
plt.plot([1, 2, 3 ,4], [10, 40 ,20 ,30])
plt.show()
```

```python
import matplotlib.pyplot as plt
#한글 오류 해결
plt.rc('font', family='Malgun Gothic')
plt.title("시험 성적 그래프", pad=10, loc="right") # pad(패딩) - 여백, 10->10px
plt.plot([1, 2, 3 ,4], [90, 80 , 85, 70])
plt.xlabel("학생번호")
plt.ylabel("시험점수")

# 격자 형태
axis = {'both', 'x', 'y'}
plt.grid(True, axis = 'x')
plt.show()
```

```python
import matplotlib.pyplot as plt
#한글 오류 해결
plt.rc('font', family='Malgun Gothic')
plt.title("범례 및 색상 표시")
plt.plot([10, 20, 30, 40], label = "비만율", color = "r")
plt.plot([40, 30, 20, 10], label = "출생율", color = "g")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
#한글 오류 해결
plt.rc('font', family='Malgun Gothic')
plt.title("마커 표시")
plt.plot([10, 20, 30, 40], 'b.', label="circle")
plt.plot([40, 30, 20, 10], 'r^', label="triangle")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [1, 2, 3, 5, 6, 7]

plt.bar(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
y=range(3)
x = [100, 400, 900]

plt.barh(y, x, color=['r','g','b'], height=0.6)
plt.yticks(y,['2018','2019','2020'] )
plt.show()
```
